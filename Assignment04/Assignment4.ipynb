```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function1(x):
    """
    Calculates 2*(x1^2 - x2^2)
    """
    return 2*(x[0]**2 - x[1]**2)

def function2(x):
    """
    Calculates x1^4 + x2^4 + 2*x1*x2
    """
    return x[0]**4 + x[1]**4 + 2*x[0]*x[1]

def function3(x):
    """
    Calculates (cos(x1))^2 + (sin(x2))^2
    """
    return np.cos(x[0])**2 + np.sin(x[1])**2
```

```python
step = 100
x = np.linspace(-5, 5, step)
y = np.linspace(-5, 5, step)
xx, yy = np.meshgrid(x, y)

z1 = function1([xx, yy])
z2 = function2([xx, yy])
z3 = function3([xx, yy])
```

```python
plt.figure(figsize=[20, 5])

plt.subplot(1, 3, 1)
plt.contour(x, y, np.transpose(z1), 100)
plt.colorbar()
plt.xlabel('x1')
plt.ylabel('x2')
plt.title(r'Level sets of $2x_1^2-2x_2^2$')

plt.subplot(1, 3, 2)
plt.contour(x, y, np.transpose(z2), 100)
plt.colorbar()
plt.xlabel('x1')
plt.ylabel('x2')
plt.title(r'Level sets of $x_1^4 + x_2^4 + 2x_1x_2$')

plt.subplot(1, 3, 3)
plt.contour(x, y, np.transpose(z3), 100)
plt.colorbar()
plt.xlabel('x1')
plt.ylabel('x2')
plt.title(r'Level sets of $cos^2(x_1) + sin^2(x_2)$')
plt.suptitle('Level sets of the functions', fontsize=20, y=1.05)
plt.savefig("./images/level_sets.png")

plt.show()
```
